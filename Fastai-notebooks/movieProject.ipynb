```
movie_reviews = []
#movie_reviews[0][0] = "a"
num = 0
temp_num = 0
#print(movie_ids)
for id in movie_ids:
  temp_num = 0
  try:
    url = "https://api.themoviedb.org/3/movie/+"+str(int(id))+"+/reviews?api_key=b0177ceaea3179cd922b9afd2e75dd83&language=en-US&page=1"
    response = requests.get(url)
    temp = response.json()
  except:
    movie_reviews[num].append("Movie is not in TMDB database")
  try:
    for val in temp['results']:
      movie_reviews[num].append(val[temp_num]['content'])
      temp_num = temp_num + 1
  except:
    movie_reviews[num][0].append("There are no reviews for this movie")
  num = num +1
print(movie_reviews)
```

```
import math
movie_reviews = []
num = 0
temp_num = 0
tempr = []
n = []
for id in movie_ids:
  tempr = []
  if math.isnan(id):
    ab = ["Movie is not in TMDB database"]
    movie_reviews.append(ab)
  else:
    try:
      url = "https://api.themoviedb.org/3/movie/+"+str(id)+"+/reviews?api_key=b0177ceaea3179cd922b9afd2e75dd83&language=en-US&page=1"
      response = requests.get(url)
      temp = response.json()
      if len(temp['results']) == 0:
        ab = ["No reviews on the API"]
        movie_reviews.append(ab)
      else:
        for i in range(len(temp['results'])):
          tempr.append(temp['results'][i]['content'])
        movie_reviews.append(tempr)
    except:
      n.append("Does not exist on database")
      movie_reviews.append(n)
  num = num +1
print(movie_reviews)
```
